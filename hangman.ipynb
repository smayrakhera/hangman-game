```python
import random

# Word list for the game
word_list = ['python', 'programming', 'openai', 'hangman', 'challenge', 'keyboard', 'laptop',"loop", "code", "java", "bash", "html", "byte", "list", "bool",
    "node", "data", "json", "git", "api", "bug", "machine", "coding", "network"]

# Pick a random word
chosen_word = random.choice(word_list)
word_length = len(chosen_word)

# Game variables
display = ['_'] * word_length
lives = 6
guessed_letters = []

print("üéÆ Welcome to Hangman!")
print("Guess the word, one letter at a time.")

# Hangman stages (empty first, builds up as lives decrease)
stages = [
    '''
      +---+
      |   |
          |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    =========
    '''
]

# Game loop
while lives > 0 and '_' in display:
    print('\n' + ' '.join(display))
    guess = input("Guess a letter: ").lower()

    if guess in guessed_letters:
        print(f"‚ö†Ô∏è You already guessed '{guess}'. Try another letter.")
        continue

    guessed_letters.append(guess)

    if guess in chosen_word:
        for i in range(word_length):
            if chosen_word[i] == guess:
                display[i] = guess
        print("‚úÖ Good guess!")
    else:
        print(f"‚ùå Wrong! You lose a life. ({lives - 1} left)")
        print(stages[6 - lives])
        lives -= 1

# Final hangman stage
print(stages[6 - lives])
print('\n' + ' '.join(display))

# Game over message
if '_' not in display:
    print(f"\nüéâ Congratulations! You guessed the word: {chosen_word}")
else:
    print(f"\nüíÄ Game Over! The word was: {chosen_word}")
```
